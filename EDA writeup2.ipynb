**np.where()**

np.where will take 3 argument values

Condition : con=data['Col1']>2

If that condition is True will provide the value:100

If that condition is False will keep the same value: data['Col1']

np.where(,,)

```python
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
q1=np.percentile(bank_df["balance"],25)
q2=np.percentile(bank_df["balance"],50)
q3=np.percentile(bank_df["balance"],75)
iqr=q3-q1
upper_bound=q3+1.5*iqr
lower_bound=q1-1.5*iqr
con1=bank_df["balance"]>upper_bound
con2=bank_df["balance"]<lower_bound
con=con1|con2
q2=bank_df["balance"].median()
bank_df["balance"]=np.where(con,q2,bank_df["balance"])
plt.boxplot(bank_df["balance"],vert=False)

plt.subplot(2,2,2)
plt.hist(bank_df["balance"])
plt.show()
```

- one hot encoder means at a time only one will be ON(1/True), others are OFF(0/False)

- Supposer housing  has two unique lables

- Yes and No

- In one hot encoder it will create two new columns

- like below

|housing|yes|no|
|-----------|---------|------|
|yes|1|0|
|no|0|1|


**Advantage**

- One hot encoder new columns are orthogonal each other

- Orthogonality means 90 degress phase shift

- 90 degress phase shift means perpendicular each other

- Perpendicular means independent each other

- Independent means no relation

- It is very important property that is to avoid relation between varaibes

**Draw back**:

- Assume that you have 100 unique lables for a column

- So it will create 100 new columns

- To process 100 columns we require more memory,more time,more hardaware 

- This is called **Curse of Dimensionality**

```python
dict1={"name":["venkatesh","Nani","preetham",np.nan,"shashi","naresh","deeksha"],
      "age":[20,21,np.nan,23,45,32,31],
      "city":[np.nan,"hyd","goa","tamilnadu","hyd","ibp","wrangal"]}
data=pd.DataFrame(dict1)
mean_value=data["age"].mean()
data["age"].fillna(mean_value,inplace=True)
data
```
